```python
def predict_house_price_features(
    Square_Footage=1500.0,
    Num_Bedrooms=3,
    Num_Bathrooms=2,
    Year_Built=2000,
    Lot_Size=0.5,
    Garage_Size=2,
    Neighborhood_Quality='Medium'  # Assume 'Low'=0, 'Medium'=1, 'High'=2
):
    lst = []

    # Numerical features
    lst.append(Square_Footage)
    lst.append(Num_Bedrooms)
    lst.append(Num_Bathrooms)
    lst.append(Year_Built)
    lst.append(Lot_Size)
    lst.append(Garage_Size)

    # Categorical encoding for Neighborhood_Quality
    quality_map = {'Low': 0, 'Medium': 1, 'High': 2}
    encoded_quality = quality_map.get(Neighborhood_Quality, 1)  # Default to 'Medium'
    lst.append(encoded_quality)

    return lst
```

```python
features = predict_house_price_features(
    Square_Footage=2100.0,
    Num_Bedrooms=4,
    Num_Bathrooms=3,
    Year_Built=2015,
    Lot_Size=0.8,
    Garage_Size=2,
    Neighborhood_Quality='High'
)

# Scale and predict
scaled_features = sc.transform([features])
predicted_price = house_model.predict(scaled_features)
print("Predicted House Price:", predicted_price)
```
